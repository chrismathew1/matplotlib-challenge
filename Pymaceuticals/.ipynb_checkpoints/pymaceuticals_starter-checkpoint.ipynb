## Observations and Insights 

# Observation 1: There is roughly an equal percentage of male and female mice: 50.4% vs 49.6%

# Observation 2: Final tumor volumes for mice under Capomulin and Rumicane regimens is generally less than the final tumor volumes for mice under Infubinol and Ceftamin. This is one indication that Capomulin and Rumicane may be more effective at reducing tumor volume.

# Observation 3: There is a moderate correlation between mouse weight and tumor volume (CC = 0.84). The heavier the mouse, the bigger the tumor. 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

#print(mouse_metadata.head())
#print(study_results.head())

# Combine the data into a single dataset
mouse_study_combined = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])

# Display the data table for preview
mouse_study_combined.head()
```

```python
# Checking the number of mice.

mice_number = len(mouse_study_combined["Mouse ID"].unique())

print(mice_number)

```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_id = pd.DataFrame(mouse_study_combined[mouse_study_combined.duplicated(["Mouse ID", "Timepoint"])]["Mouse ID"])

duplicate_id = pd.DataFrame(duplicate_id["Mouse ID"].unique())

duplicate_id
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse = mouse_study_combined.loc[mouse_study_combined["Mouse ID"] == "g989"]

duplicate_mouse
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
combined_clean = mouse_study_combined.loc[mouse_study_combined["Mouse ID"] != "g989"]

combined_clean
```

```python
# Checking the number of mice in the clean DataFrame.
mice_number = len(combined_clean["Mouse ID"].unique())
print(mice_number)
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor 
# volume for each regimen



# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

means = combined_clean.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
medians = combined_clean.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
variances = combined_clean.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
stds = combined_clean.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]
sems = combined_clean.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]

summary_stats = pd.DataFrame({
                                "Mean Tumor Volume": means,
                                "Median Tumor Volume": medians,
                                "Variance Tumor Volume": variances,
                                "Std Tumor Volume": stds,
                                "SEM Tumor Volume": sems
                                
    
})


summary_stats

```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and 
# SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line

summary_stats_agg = combined_clean.groupby("Drug Regimen").agg({"Tumor Volume (mm3)" : ["mean", "median", "var", "std", "sem"]})

summary_stats_agg
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

max_df = combined_clean.groupby(["Mouse ID"])['Timepoint'].max()

max_df = max_df.reset_index()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint

merged = pd.merge(max_df, combined_clean,on=['Mouse ID', 'Timepoint'])
```

```python
# Put treatments into a list for for loop (and later for plot labels)

treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]


# Create empty list to fill with tumor vol data (for plotting)

tumor_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

for treatment in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    drug_vol = merged["Tumor Volume (mm3)"].loc[(merged["Drug Regimen"] == treatment)]
    
    
    # add subset 
    tumor_data.append(drug_vol)
    
    # Determine outliers using upper and lower bounds
    quartiles = drug_vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq

    print(f"The lower quartile of tumor volume for {treatment} is: {lowerq}")
    print(f"The upper quartile of tumor volume for {treatment} is: {upperq}")
    print(f"The interquartile range of tumor volume for {treatment} is: {iqr}")
    print(f"The the median tumor volume for {treatment} is: {quartiles[0.5]} ")

    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    print(f"Values below {lower_bound} could be outliers.")
    print(f"Values above {upper_bound} could be outliers.")
    print()
    
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

green_diamond = dict(markerfacecolor='b', marker='D')
plt.boxplot(tumor_data, labels=treatments, flierprops=green_diamond)
plt.title("Box plot of final tumor volume across four drug regimens")
plt.ylabel("Final tumor volume (mm3)")
plt.show
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
scatter_df = combined_clean.loc[combined_clean["Drug Regimen"] == "Capomulin"]
scatter_df.head()

scatter_df = scatter_df.groupby("Mouse ID").mean()

plt.scatter(scatter_df["Weight (g)"], scatter_df["Tumor Volume (mm3)"], marker="o", facecolors="red", edgecolors="black")
plt.title("Average Tumor Volume vs. Mouse Weight")
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
cc = round(st.pearsonr(scatter_df["Weight (g)"], scatter_df["Tumor Volume (mm3)"])[0],2)
print(f"The correlation coefficient between average tumor volume and mouse weight is {cc}")

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(scatter_df["Weight (g)"], scatter_df["Tumor Volume (mm3)"])
regress_values = scatter_df["Weight (g)"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(scatter_df["Weight (g)"], scatter_df["Tumor Volume (mm3)"], marker="o", facecolors="red", edgecolors="black")
plt.plot(scatter_df["Weight (g)"],regress_values,"b-")
plt.annotate(line_eq,(20,40),fontsize=15,color="red")
plt.title("Average Tumor Volume vs. Mouse Weight")
plt.xlabel("Mouse Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
print(f"The r-squared is: {rvalue**2}")
plt.show()
```
